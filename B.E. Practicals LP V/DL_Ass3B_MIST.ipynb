````python
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow import keras
import numpy as np

# PROJECT: Multiclass classification on Fashion-MNIST dataset
# OBJECTIVE: Build a CNN to classify clothing items into 10 categories
# DATA: 70,000 grayscale images (28x28) of fashion items labeled 0-9:
#   0 T-shirt/top, 1 Trouser, 2 Pullover, 3 Dress, 4 Coat,
#   5 Sandal, 6 Shirt, 7 Sneaker, 8 Bag, 9 Ankle boot

# -------------------- Load the Dataset ---------------------
(x_train, y_train), (x_test, y_test) = keras.datasets.fashion_mnist.load_data()

# Display two sample images
plt.figure(figsize=(4,2))
plt.subplot(1,2,1)
plt.imshow(x_train[0], cmap='gray')
plt.title(f"Label: {y_train[0]}")
plt.axis('off')
plt.subplot(1,2,2)
plt.imshow(x_train[1], cmap='gray')
plt.title(f"Label: {y_train[1]}")
plt.axis('off')
plt.show()

# -------------------- Preprocess Data ---------------------
# Convert to float32 and scale pixel values to [0,1]
x_train = x_train.astype('float32') / 255.0
x_test  = x_test.astype('float32')  / 255.0

# Reshape images to include channel dimension
# -1 lets NumPy infer the sample count (60000 or 10000)
# Final shape: (samples, 28, 28, 1)
x_train = x_train.reshape(-1, 28, 28, 1)
x_test  = x_test.reshape(-1, 28, 28, 1)

print("x_train shape:", x_train.shape)
print("x_test shape:",  x_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:",  y_test.shape)

# -------------------- Build the Model ---------------------
# Convolutional Neural Network architecture:
# - Conv2D layers extract spatial features
# - MaxPooling2D downsamples spatial dimensions
# - Dropout prevents overfitting by randomly dropping neurons
# - Flatten converts 3D feature maps to 1D vectors
# - Dense layers perform classification

model = keras.Sequential([
    keras.layers.Conv2D(32, (3,3), activation='relu', input_shape=(28,28,1)),
    # 32 filters, each 3x3. Output shape: (26,26,32)
    keras.layers.MaxPooling2D((2,2)),
    # Reduces spatial dims to (13,13,32)
    keras.layers.Dropout(0.25),  # 25% dropout

    keras.layers.Conv2D(64, (3,3), activation='relu'),
    # 64 filters -> output (11,11,64)
    keras.layers.MaxPooling2D((2,2)),
    # Output (5,5,64)
    keras.layers.Dropout(0.25),

    keras.layers.Conv2D(128, (3,3), activation='relu'),
    # 128 filters -> output (3,3,128)

    keras.layers.Flatten(),
    # Flattens to vector of size 3*3*128=1152

    keras.layers.Dense(128, activation='relu'),
    # Fully connected layer with 128 neurons
    keras.layers.Dropout(0.25),

    keras.layers.Dense(10, activation='softmax')
    # Output layer: 10 classes, softmax for probability distribution
])

model.summary()

# -------------------- Compile the Model --------------------
# optimizer='adam': adaptive learning rate optimizer
# loss='sparse_categorical_crossentropy': suitable when labels are integers 0-9
# metrics=['accuracy'] tracks fraction of correct predictions
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# -------------------- Train the Model ----------------------
# epochs: number of full passes through training data
# validation_data: evaluates performance on test set each epoch
history = model.fit(
    x_train, y_train,
    epochs=10,
    validation_data=(x_test, y_test)
)

# -------------------- Evaluate the Model -------------------
test_loss, test_acc = model.evaluate(x_test, y_test, verbose=2)
print(f"Test accuracy: {test_acc:.4f}")

# ------------------- Plot Training Curves ------------------
# Accuracy over epochs
plt.figure()
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Model Accuracy over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()

# Loss over epochs
plt.figure()
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()
````

## 4. Preprocessing

```python
x_train = x_train.astype('float32') / 255.0
x_test  = x_test.astype('float32')  / 255.0

x_train = x_train.reshape(-1, 28, 28, 1)
x_test  = x_test.reshape(-1, 28, 28, 1)
```

1. **`.astype('float32')`**

   * Converts integer pixel data to 32‑bit floats.
   * **Why?** Neural nets expect floats and 32‑bit is GPU‑friendly.

2. **`/ 255.0`**

   * Scales pixel values from \[0,255] → \[0,1].
   * **Why?** Normalizing inputs helps the network converge more quickly and stably.

3. **`.reshape(-1, 28, 28, 1)`**

   * Changes array shape to `(num_samples, height, width, channels)`.
   * **`-1`** tells NumPy to infer the sample count.
   * The final `1` channel dimension is needed by `Conv2D`.

---

## 5. Building the CNN

```python
model = keras.Sequential([
    keras.layers.Conv2D(32, (3,3), activation='relu', input_shape=(28,28,1)),
    keras.layers.MaxPooling2D((2,2)),
    keras.layers.Dropout(0.25),

    keras.layers.Conv2D(64, (3,3), activation='relu'),
    keras.layers.MaxPooling2D((2,2)),
    keras.layers.Dropout(0.25),

    keras.layers.Conv2D(128, (3,3), activation='relu'),

    keras.layers.Flatten(),
    keras.layers.Dense(128, activation='relu'),
    keras.layers.Dropout(0.25),

    keras.layers.Dense(10, activation='softmax')
])
model.summary()
```

1. **`Sequential([...])`**

   * A simple container: layers are applied in order.

2. **`Conv2D(filters, kernel_size, activation, input_shape)`**

   * **`filters=32`**: number of feature detectors.
   * **`kernel_size=(3,3)`**: each filter is 3×3 pixels.
   * **`activation='relu'`**: applies non‑linearity $\max(0,x)$.
   * **`input_shape=(28,28,1)`**: only on the first layer—tells Keras what shape each example has.

3. **`MaxPooling2D(pool_size)`**

   * **`pool_size=(2,2)`**: downsamples each feature map by taking the max over non‑overlapping 2×2 patches.
   * **Why?** Reduces spatial dimensions, computation, and helps the model focus on the strongest activations.

4. **`Dropout(rate)`**

   * **`rate=0.25`**: randomly zeroes 25% of the inputs each update.
   * **Why?** A regularizer to prevent overfitting by forcing the network to learn redundant representations.

5. **`Flatten()`**

   * Converts the 3D tensor output by the last convolution into a 1D vector for the Dense layers.

6. **`Dense(units, activation)`**

   * **Hidden Dense layer:** `Dense(128, activation='relu')` learns high‑level combinations of the convolutional features.
   * **Output layer:** `Dense(10, activation='softmax')` outputs a length‑10 probability vector summing to 1 over the ten classes.

7. **`model.summary()`**

   * Prints each layer’s name, output shape, and number of parameters for easy inspection.

---

## 6. Compiling the Model

```python
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

1. **`optimizer='adam'`**

   * An adaptive optimizer that adjusts learning rates per‑parameter using estimates of first and second moments of the gradients.

2. **`loss='sparse_categorical_crossentropy'`**

   * Suitable for multiclass integer labels (0–9) with a softmax output.
   * Equivalent to one‑hot `categorical_crossentropy`, but takes integer targets directly.

3. **`metrics=['accuracy']`**

   * Tells Keras to compute the fraction of correctly classified samples during training/validation.

---

## 7. Training

```python
history = model.fit(
    x_train, y_train,
    epochs=10,
    validation_data=(x_test, y_test)
)
```

* **`x_train, y_train`**: training data.
* **`epochs=10`**: number of full passes over the training set.
* **`validation_data=(x_test, y_test)`**: after each epoch, compute loss & metrics on the test set to monitor generalization.

The `fit()` call returns a **History** object that stores the training/validation loss and accuracy per epoch.

---

## 8. Evaluation

```python
test_loss, test_acc = model.evaluate(x_test, y_test, verbose=2)
print(f"Test accuracy: {test_acc:.4f}")
```

* **`model.evaluate(x, y)`**

  * Computes the final loss and specified metrics on the provided dataset.
  * **`verbose=2`** prints a brief summary.

# Accuracy

* Plotting them lets you see:

  * **Convergence:** loss should decrease, accuracy increase.
  * **Overfitting:** if validation curves diverge (e.g. validation loss rises), the model is memorizing training data.

---

### Why This Pipeline?

1. **Normalization & Reshaping:** Essential for stable, fast training and compatibility with convolutional layers.
2. **Convolution + Pooling:** Learns spatial patterns (edges, textures) and reduces data dimensionality.
3. **Dropout:** Regularizes the model, combating overfitting on a relatively small dataset.
4. **Dense + Softmax:** Translates extracted features into class probabilities.
5. **`adam` + crossentropy:** A proven optimizer/loss combo for classification tasks.
6. **Visualization:** Quickly diagnose training behavior and spot under‑ or over‑fitting.

Together, these functions form the canonical workflow for image‑based multiclass classification using Keras.
